# Data Preprocessing

```python
df = df[df['Final Profit Margin'] > 0.5]
```

```python
plt.hist(df['Final Profit Margin'], bins=5, edgecolor='black')
```

```python
# Transfer State to 'East coast' and 'West coast'
import pandas as pd

df['State'] = df['State'].str.title()

east_criteria = df['State'].str.startswith(('Connecticut','Delaware','Florida','Maine','Maryland','Massachusetts','Mississippi','New Hampshire','New Jersey','New York','North Carolina','Pennsylvania','Rhode Island','South Carolina','Vermont','Virginia','West Virginia'))
west_criteria = df['State'].str.startswith(("Alabama","Alaska", "Arizona", "Arkansas", "California", "Colorado", "Georgia","Indiana","Kentucky"
    "Hawaii", "Idaho", "Illinois", "Iowa", "Kansas",
    "Louisiana","Michigan", "Minnesota", "Missouri", "Montana", "Nebraska",
    "Nevada", "New Mexico", "North Dakota","Ohio", "Oklahoma", "Oregon",
    "South Dakota","Tennessee", "Texas", "Utah", "Washington","Wisconsin", "Wyoming"))

# Combine the criteria to create a final boolean mask
final_criteria = east_criteria | west_criteria

# Drop rows with NaN values in 'State' column
df = df.dropna(subset=['State'])

# Create a mask for non-NaN values
not_nan_mask = ~df['State'].isna()

# Apply the final criteria and the not-NaN mask
df['State'] = df['State'].mask(final_criteria & not_nan_mask, 1).mask(west_criteria & not_nan_mask, 0)

df['State'] = df['State'].apply(lambda x: 'west' if x == 0 else 'east')
df.head()
```

# Regression Model (Final Profit Margin)

```python
# Dependent: Final Profit Margin
# Independent: Customer Type, Material Type - dummies

X = df[['Customer Type','Material Type']]
Y = df[['Final Profit Margin']]

# Convert categorical variable into dummy/indicator variables and drop one in each category
X = pd.get_dummies(data=X, drop_first=True)
```

```python
# check what x variables look like after dummies
X.head()
```

```python
import pandas as pd
import statsmodels.api as sm

# Adding a constant to the model (for intercept)
X = sm.add_constant(X)

#fitting the repression model
model = sm.OLS(Y,X).fit()

print(model.summary())
```

# Regression model (Quantity)

```python
data_q = pd.get_dummies(df, columns=['State', 'Customer Type'], drop_first=True)
data_q.head()
```

```python
data_q.info()
```

```python
data_q["Unit Price"] = pd.to_numeric(data_q["Unit Price"], errors='coerce')
```

```python
import pandas as pd
import statsmodels.api as sm

# Define independent variables (X) and dependent variable (y)
X = data_q[['State_west', 'Customer Type_Non Member', 'Unit Price']]
y = data_q['Quantity']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```
